```python
from bs4 import BeautifulSoup
import requests
page_link='http://stats.espncricinfo.com/indian-premier-league-2016/engine/records/averages/batting.html?id=11001;type=tournament'
page_response=requests.get(page_link,timeout=5)
page_content=BeautifulSoup(page_response.content,"html.parser")
textcontent=[]
```

```python
soup = BeautifulSoup(page_response.content, 'html5lib')
table=soup.findAll('td',attrs={'nowrap':'nowrap'})
```

```python
print(table)
```

```python
import csv
csvfile=open('/home/srikar/Desktop/Big_Data_Project/Phase_1_Final_batsman.csv', 'w', newline='')
counter=0
row=list()
row_list=list()
index=0
spamwriter = csv.writer(csvfile)
for i in table:
    if(counter%14==0 and counter!=0):
        spamwriter.writerow(row)
        row_list.append(row)
        v=row[6]
        if(v=='-'):
            v='0'
        print(str(index)+" 1: "+v,row[0])
        index+=1
        row.clear()
    val=i.text
    row.append(val)
    counter+=1
```
